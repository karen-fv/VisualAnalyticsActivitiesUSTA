# Visualizaciones de Datos

# Covid 19

## Introducción

Para nadie es un secreto que el Covid-19 es una de las pandemias que ha cobrado miles de vidas en el último año alrededor del mundo, este virus surgió a finales del año 2019 en Wuhan China en un mercado de animales donde inició su propagación en diferentes partes del mundo.

Para los científicos de datos el tema del Covid 19 ha sido tendencia en la creación de nueva visualizaciones sobre la expansión del virus, puesto que de esta situación se tiene un montón de registros y datos a nivel mundial que permiten a los creadores de las mismas poner su creatividad a prueba.Pero en este documento se describirán  dos visualización en particular que fueron publicadas  el 24 de abril de 2020 por el profesor y director Lluís Vicent en la página de Barcelona School of management,  una pretendía explicar en un minuto la expansión que tenía el virus hacia Estados Unidos y la otra mostraba la letalidad del virus en los diferentes paises.




### Visualización de infectados y letalidad de Covid 19 por países
Por medio de un gráfico iterativo de puntos se visualiza la evolución del Covid 19  hasta el dia 107 con respecto a la letalidad y cantidad de infectados por cada país en estudio. A través de un plano cartesiano se representa en el eje X la letalidad y en el eje Y la cantidad de infectados, los punto representan los países y el tamaño de ellos es proporcional a la cantidad de infectados por país, los colores están asignados por continente.
 
***Colores***
1. Azúl= Europa
2. Rojo = América
3. Naranja = Asia
4. Verde = Africa
5. Turquesa = Oceanía

## Comentarios y críticas
1. Esta visualización de puntos fue útil para comprender  la magnitud de la situación vivida por el covid 19 al rededor del  mundo en los 107 días analizados.
2. Es una representación eficiente, sencilla de entender y agradable a la vista del público.
3. Esta visualización de puntos representa un dataset availability dynamic, que tiene en cuenta datos como la letalidad del virus y la cantidad de contagios , este tipo de datos serían atributos  de orden secuencial partiendo de 0 en adelante,  ya que en un país no se reportan -Y casos de contagio o -X casos letales, estos datos se pueden agrupar en un tabla de reportes por país , además el gráfico se realiza sobre un plano cartesiano.
4. El objetivo de esta visualización es analizar y presentar a los usuarios los paises con mas contagio y letalidad, además presenta un target conocido como Lookup debido a que se conoce los datos y localización de los mismos , también es un gráfico que sirve para identificar lo que sucede con el paso de los días y es un comparativo clasificado en All data Outliers.
5. La forma de organizar los datos graficamente es un separate puesto que, los datos se separan en puntos y colores, en este caso los colores representan cada continente y los puntos cada país.
6. La marca que maneja esta visualización es de puntos y sus canales son de color y area “Size+hue(color)”.



### Visualización expansión Covid 19  desde China hasta Estados Unidos 

A través de un mapamundi iterativo se visualiza en un minuto la propagación del virus desde China hasta Estados Unidos día a día hasta llegar hasta el día 107 de propagación. Por medio de colores se pueden identificar los infectados por cada 100.000 habitantes en los diferentes países afectados, por otro lado se encuentra el grosor de los puntos negros en la visualización que significa la cantidad acumulada infectados en cada pais.

## Comentarios y críticas

1. Esta visualización del mapamundi es muy eficiente, didáctica  y útil a la hora de mostrar la velocidad de propagación del virus a nivel mundial, aunque su objetivo principal es representar la velocidad de propagación desde china hasta Estados unidos, pero sin embargo representa la expansión del covid en los demás países.
2. Esta gráfica pertenece a un Dataset de tipo geométrico espacial , porque tiene en cuenta datos como :  el país, la cantidad de contagiados, día de registro  de infectados, y la acomulación de contagios que deja en cada país, lo que significa que tiene en cuenta los Ítems y la posición .
3. La representación tiene el objetivo de mostrar a los usuarios la velocidad de contagio desde China hasta  EE.UU, posee un target conocido como Lookup asociado también a Shape, que es el target de datos espaciales.
4. La visualización iterativa organiza los datos de acuerdo a la ubicación del país (Arrange-Use), También  utiliza algunos colores y puntos para identificar diferencias entre los  atributos de cada país.
5. La marca que maneja este gráfico es  de áreas debido a las fronteras que existen entre países y el canal es de  color y posición “Position+Hue(color)”.
