```python
# ------- Dirigido Ej #1
plt.figure(figsize=(7,7))
# Construir un data frame para ver las conexiones
df = pd.DataFrame({ 'from':['Inicio','Inicio','Inicio','A','B','D','E','C','G','F','H','G','J','I'], 'to':['A','B','C','D','E','G','F','F','J','H','I','I','Final','Final'],'value':[1,1,1,5,5,5,14,10,22,18,14,26,18,26]})
 
# Contruir el grafo 
G=nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph())
 
nx.draw(G, with_labels=True, 
        node_color='RosyBrown', 
        node_size=1700, 
        edge_color=df['value'], 
        width=5, 
        edge_cmap=plt.cm.Wistia)    
    
# Imprimir el grafo
#nx.draw(G, with_labels=True)
#plt.title("Dirigido")
#plt.show()
```

# Framework 

#### What?
Esta gráfica corresponde a un data set tipo redes (Networks) con nodos y atributos, utiliza tres variable cualitativas,  una secuencial que es la actividad representada en las letras del alfabeto desde la A hasta la  J, por otro lado tenemos las descripciones de cada actividad que corresponde a una variable cualitativa nominal y por ultimo tenemos una variable cualitativa ordinal que es precedente inmediatos lo que indica el orden y la relación que se debe seguir para llevar a cabo la conferencia. Cabe destacar que una visualización estática. 

#### Why?
Esta visualización pretende presentar y explicar la relación de conexión y dependencia entre las actividades necesarias para llevar a cabo el seminario de capacitación en ventas de **Bluebird University** por medio de la representación de la ruta *( Paths)*  a través de   un grafo dirigido *(DIAGRAMA PERT-CPM)*. 

#### How?
Esta red es un grafo dirigido mas conocido como diagrama **pert-cpm**, el cual identifica las actividades y las organiza de acuerdo con la relación entre sí, como lo muestra la variable de precedentes inmediatos que serían las actividades predecesoras, teniendo en cuenta una especie de algoritmo que permite mostrar la ruta (Paths) a seguir para llegar al objetivo final.

### Marcas y canales
Esta visualización utiliza marca de líneas para los enlaces dirigidos que representan la relación y marca de puntos para los diez nodos más el nodo de inicio y el de final, que representan las tareas a realizar que en este caso son las actividades, pero la tarea principal de este modismo es localizar caminos. Este modismo utiliza colores en los enlaces para identificar más fácil la relación entre las actividades, es decir entre más claro es el enlace significa que esta empezando pero a medida que avanza el color se va haciendo mas y mas fuerte indicando que se acerca al final del camino.

### Reglas generales descritas 
El grafico trata de cumplir es su mayoría las reglas generales puesto que no presenta problemas de oclusión, tampoco se justifica utilizar un 3D ni un 2D, ya que, así como esta se observa claramente la ruta sin distorsión alguna, también hace buen uso del espacio y resolución de pixeles.
